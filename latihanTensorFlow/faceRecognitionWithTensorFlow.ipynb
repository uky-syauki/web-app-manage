```python
import matplotlib.pyplot as plt
import numpy as np

# Data untuk diagram batang
categories = ['Kategori 1', 'Kategori 2', 'Kategori 3']
values1 = [10, 15, 7]
values2 = [5, 12, 9]
values3 = [8, 10, 11]

# Menghasilkan indeks untuk posisi batang
x = np.arange(len(categories))

# Lebar batang
bar_width = 0.2

# Membuat diagram batang
plt.bar(x - bar_width, values1, width=bar_width, label='Bar 1')
plt.bar(x, values2, width=bar_width, label='Bar 2')
plt.bar(x + bar_width, values3, width=bar_width, label='Bar 3')

# Menambahkan label kategori
plt.xlabel('Kategori')
plt.ylabel('Nilai')

# Menambahkan label batang
plt.xticks(x, categories)

# Menambahkan legenda
plt.legend()

# Menampilkan diagram batang
plt.show()
```

```python
# Impor library yang diperlukan
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Memuat dataset Iris
iris = datasets.load_iris()
X = iris.data  # Fitur-fitur
y = iris.target  # Label

# Membagi dataset menjadi data pelatihan dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi model klasifikasi (K-Nearest Neighbors)
knn = KNeighborsClassifier(n_neighbors=3)

# Melatih model dengan data pelatihan
knn.fit(X_train, y_train)

# Melakukan prediksi menggunakan data uji
y_pred = knn.predict(X_test)

# Mengukur akurasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi Model: {accuracy * 100:.2f}%')
```
